# Conteo de fijaciones y movimientos sacÃ¡dicos

```python
import csv
with open('GF2pupilas.csv') as file_obj:
    
    reader_obj = csv.reader(file_obj)
    reader_objIJ = list(reader_obj) 
print(reader_objIJ[1][12])
```

```python
count = 0
for i in range (1,r-1):
    if reader_objIJ[i][12] == reader_objIJ[1][12]: # [1][12] = fijacion
        count +=1
percentage1 = round((count*100)/(r-1),2)
print('La cantidad de fijaciones reportadas es de: ',count,'que corresponde al',percentage1,'%')

countt = 0
for i in range (1,r-1):
    if reader_objIJ[i][12] == reader_objIJ[4][12]: #sacadifos
        countt += 1
percentage2 = round((countt*100)/(r-1),2)
print('La cantidad de movimientos sacÃ¡dicos reportados es de: ',countt,'que corresponde al',percentage2,'%')


counttt = 0
for i in range (1,r-1):
    if reader_objIJ[i][12] == reader_objIJ[10][12]: #no clasificado
        counttt += 1
percentage3 = round((counttt*100)/(r-1),2)
print('La cantidad de movimientos no clasificados es de: ',counttt,'que corresponde al',percentage3,'%')

print('Los datos se limpiaron del resultado "EyesNotFound"')
        
```

```python
x = np.linspace(700, 1000, 11)
y = np.linspace(500, 1000, 11)
print('x = ',x,)
print('y = ',y,)
len(x)
```

```python
x[3]
#type(x[3]) is float
```

```python
#gazeX[][] = reader_objIJ[1][6] = [n]][6]    Gaze point X
#gazeY[][] = reader_objIJ[1][7] = [n][7]   Gaze point Y
gazeX,gazeY = 750, 900 #iloc[[n][6/7]]
p #= 50982 

HM = np.zeros(len(x),len(y))

# FALTA EL AVANCE PARA CADA FILA del par de datos (x,y)
for n in range (len(p-1)): 
    for i in range(len(x-1)):
        if gazeX[n][6] =< x[i]: # si ya es menor procedo a comparar a GazeY
            for j in range(len(y-1)):
                if gazeY[n][7] =< y[j]: # si ya es menor guardamos la coordenada en el HM
                    HM[x[i]][y[j]] =+ 1
            
print(sns.heatmap(HM))        
        
```
